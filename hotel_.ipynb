```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36 Edg/116.0.1938.54'}
source = requests.get('https://www.booking.com/searchresults.en-gb.html?ss=Vijayaw%C4%81da%2C+India&ssne=Hyderabad&ssne_untouched=Hyderabad&label=msn-HiTydpHndf_KligNqm9Sgw-79852220055838%3Atikwd-79852393960302%3Aloc-90%3Aneo%3Amte%3Alp149155%3Adec%3Aqsbooking.com&aid=2369661&lang=en-gb&sb=1&src_elem=sb&src=searchresults&dest_id=-2114438&dest_type=city&checkin=2024-10-09&checkout=2024-10-10&group_adults=2&no_rooms=1&group_children=0',headers=headers)
```

```python
source.raise_for_status()
soup = BeautifulSoup(source.text,'html.parser')
```

```python
hotels = soup.find_all('div',class_="b1037148f8 c730b02848")
ratings = soup.find_all('div',class_="a3b8729ab1 d86cee9b25")
price = soup.find_all('span',class_="f6431b446c fbfd7c1165 e84eb96b1f")
numberofreviews = soup.find_all('div',class_="abf093bdfe f45d8e4c32 d935416c47")
```

```python
numberofreviews
```

```python
hs=[]
ra=[]
pr=[]
nr=[]
for hotel in hotels:
    hs.append(hotel.text)
for rating in ratings:
    r=rating.text
    ra.append(float(r.split(' ')[1]))
for prices in price:
    pr.append(prices.text)
for numberofreview in numberofreviews:
    nr.append(numberofreview.text)
```

```python
print(hs)
```

```python
print(pr)
```

```python
print(ra)
```

```python
print(nr)
```

```python
pattern = r'\d{1,3}(?:,\d{3})*'
price_for_onenight_stay = [int(re.search(pattern, value).group().replace(',', '')) for value in pr]
price_for_onenight_stay
```

```python
print(len(hs))
print(len(ra))
print(len(price_for_onenight_stay))
print(len(nr))
```

```python
d={"hotel_name":hs,"Rating":ra,"price_for_onenight_stay":price_for_onenight_stay,"number_of_reviews":nr}
d
```

```python
model=pd.DataFrame(d)
model
```
